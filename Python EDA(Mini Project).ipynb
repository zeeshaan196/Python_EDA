```python
# SEABORN PLOTTING FOR LINE , SCATTER , HISTOGRAM
CD = A.groupby(['Maker','Gearbox','Bodytype'])['Width'].value_counts().reset_index().sort_values(by = 'Width' ,ascending = False).head(30)
CD
```

```python
# SEABORN PLOTTING FOR LINE , SCATTER , HISTOGRAM
DE = A.groupby(['Maker' , 'Gearbox','Bodytype'])['Length'].value_counts().reset_index().sort_values(by = 'Length' , ascending = False).head(30)
DE
```

```python
# MERGING 2 DATA's 
EF = pd.merge(CD , DE, on = 'Maker' , how = 'inner')
EF
```

```python
CD.columns
```

```python
plt.figure(figsize = (12,5))
sns.lineplot(x = 'Maker' , y = 'Width', data = CD  , lw = 3 , label = 'Width of Makers' , marker = 'X' , markersize = 9 )
sns.lineplot(x = 'Maker' , y = 'Length' , data = DE  , lw = 3 , label = 'Length of Makers' , marker = "o" , markersize = 8)
sns.lineplot(x = 'Maker' , y = 'count_x', data = EF  , lw = 5 , label = 'Count_x of Makers After Merge' , marker = '*' , markersize = 15 )
sns.lineplot(x = 'Maker' , y = 'count_y', data = EF  , lw = 5 , label = 'Count_y of Makers After Merge' , marker = '*' , markersize = 15 )
plt.xlabel('MAKERS' , color = 'violet')
plt.ylabel('Width and Length' , color = 'purple')
plt.xticks(rotation = 90)
plt.yticks(rotation = 45)
plt.title('REPORT OF MAKERS ON SIZE')
plt.grid(alpha = 0.8)
plt.show()
```

```python
plt.figure(figsize = (11,5))
sns.scatterplot(x = 'Width' , y = 'count' , data = CD , label = 'Width of Cars',marker='o',s=80)
sns.scatterplot(x = 'Length' , y ='count' , data = DE , label = 'Length of Cars',marker='^',s=80)
plt.title('SCATTERED REPORT OF MAKERS BASED ON SIZE',fontsize=16, color='purple')
plt.xlabel('Width and length of CARS' , color = 'blue')
plt.ylabel('No. of Counts according to their size' , color = 'red')
plt.xticks(rotation = 90 , color = 'brown')
plt.yticks(rotation = 45 , color = 'orange')
plt.tight_layout()
plt.grid(alpha = 0.8)
plt.show()
```
